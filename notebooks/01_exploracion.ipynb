```python
import pandas as pd
import os

# 1. Validar que los archivos existan
raw_path = "../data/raw/"
archivos_esperados = [
    "clientes.csv", "provincias.csv", "segmentos.csv", "productos.csv",
    "cliente_estado_mensual.csv", "cliente_producto_mensual.csv", 
    "cliente_producto_alta.csv", "dataset_next_best_product.csv"
]

for archivo in archivos_esperados:
    completo = os.path.join(raw_path, archivo)
    if os.path.exists(completo):
        print(f"{archivo} detectado correctamente.")
    else:
        print(f"No se encuentra {archivo}")

# 2. Exploración rápida (repite este bloque por cada archivo)
df_clientes = pd.read_csv(f"{raw_path}clientes.csv")
print("\n--- Estructura de Clientes ---")
print(f"Filas y columnas: {df_clientes.shape}")
print(df_clientes.info())
print("Valores nulos detectados:")
print(df_clientes.isnull().sum())
```
