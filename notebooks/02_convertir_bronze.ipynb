```python
import pandas as pd
import os
from datetime import datetime

raw_path = "../data/raw/"
bronze_path = "../data/bronze/"

# Asegurar que la carpeta bronze exista
os.makedirs(bronze_path, exist_ok=True)

archivos = [
    "clientes", "provincias", "segmentos", "productos",
    "cliente_estado_mensual", "cliente_producto_mensual", 
]

for item in archivos:
    # 1. Leer de Raw
    df = pd.read_csv(f"{raw_path}{item}.csv")
    filas_raw = len(df)
    
    # 2. Columnas de Auditoría (Trazabilidad profesional)
    df['fecha_carga'] = datetime.now()
    df['archivo_origen'] = f"{item}.csv"
    df['capa_origen'] = "Raw"
    
    # 3. Guardar en Bronze en formato Parquet
    destino_parquet = f"{bronze_path}{item}.parquet"
    df.to_parquet(destino_parquet, index=False)
    
    # 4. Validación de consistencia de registros
    df_parquet = pd.read_parquet(destino_parquet)
    filas_parquet = len(df_parquet)
    
    if filas_raw == filas_parquet:
        print(f"{item} convertido con éxito. Registros íntegros: {filas_parquet}")
    else:
        print(f" Error de consistencia en {item}. Raw: {filas_raw} vs Parquet: {filas_parquet}")
```
